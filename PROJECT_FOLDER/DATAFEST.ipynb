```python
#Evaluating whether two categorical variables are independent.
categorical_columns = data[['Age', 'Gender', 'Ethnicity', 'Disability_status',
                               'Family_size', 'FamilyIncomeRange', 'EducationLevel',
                               'Final_grade']]

results = []
for i, col1 in enumerate(categorical_columns):
    for j, col2 in enumerate(categorical_columns):
        if i < j:  # To avoid repeating pairs (like col1, col2) and (col2, col1)

            contingency_table = pd.crosstab(data[col1], data[col2])


            chi2, p, dof, expected = chi2_contingency(contingency_table)


            results.append({
                'Variable 1': col1,
                'Variable 2': col2,
                'Chi2': chi2,
                'p-value': p,
                'Degrees of Freedom': dof
            })


results_df = pd.DataFrame(results)

results_df
```
